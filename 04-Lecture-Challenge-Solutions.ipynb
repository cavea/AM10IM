This challenge is to test your understanding of for loops. using the Collatz conjecture:

__Challenge 1__: Can you ammend the code for the Collatz conjecture so that it will also plot the maximum number that each integer reaches? For example the number 27 reaches 9232. The number 28 reaches 52. You may find the MatLab function max(x) useful

```octave
clear all
n = 1000;       
t = [];
m = [0]; %m is going to store all the maximum values of z.
for j = 1:n
    z = j;          
    zz = []; % HERE, we need to find the maximum so we need to store all iterations. Each new number it is reset
    i = 0;          
    while(z ~= 1)   
        if (mod(z,2)==0)
            zz = [zz z]; % HERE we store each iteration of z as the row vector zz
            z = z./2;         
        else
            zz = [zz z];  % And it must be done for every case, whether it was an odd number or even.
            z = 3.*z+1;       
        end
    i = i+1;
    end
    m = [m max(zz)]; %Here we have added on the maximum value of zz and stored it in m.
    t = [t i];               
end
plot(t,'*')
title('Stopping Times')
figure
plot(m,'*')
title('Maximum Values')
```

__Challenge 2__ Can you plot the stopping time of the numbers with a stopping time larger than all smaller initial starting values?

We have stored all the stopping times, we now have to check which stopping times are greatest than all the ones previously:

```octave
tt = [1 0]; % this will what we looking for, initialised with stopping time of 1, which is zero
for i = 2:n
    if t(i) > max(tt(:,2)) % If the i'th stopping time is greater than the maximum value in tt
        tt = [tt; i t(i)]; % then add a new row to tt, i and the new 'maximum'
    end
end
plot(tt(:,1),tt(:,2),'*')
```

Interestingly we could do this with the maximum value that is reached for each number and we can see that the numbers that take the longest time, do not reach the highest values.

```octave
mm = [1 1];
for i = 2:n
    if m(i) > max(mm(:,2))
        mm = [mm; i m(i)];
    end
end
mm
```

__Challenge 3__ Consider the following game of threes. Pick a positive integer. If it is divisble by three, divide by three. If it is not either add or subract 1 (whichever makes it divisible by three). Repeat. Can you compare the stopping times for all the integers from 1 to 1000?

This works very similar to our Collatz conjecture, in fact this game is guarenteed to reach 1. Let's calculate the stopping distances:

```octave
n = 1000;
tt =[];                     % tt will store all the stopping times
for i = 1:n                 % This for loop will loop from 1 to 1000 
  z = i;
  t = 0;                    % This will be the stopping time, it is initialised outside the while loop
  while (z ~= 1)            % and inside the for loop.
    if mod(z,3) == 0 
      z = z./3;             % if z is divisible by 3, divide by 3
    elseif mod(z,3) == 1
      z = z-1;              % elseif z is remainder 1, take away 1
    else
      z = z+1;              % else (only other possibility) add 1
    end
    t = t+1;                % whatever has happened, increment the time by 1
  end
  tt = [tt t];
end
plot(tt,'*')
title('Stopping Times for "Game of Threes"')
```
