__Challenge 1__ Create a script that will generate a tridiagonal matrix of size $50\times 50$, where $\alpha = 0.1$

```octave
n = 50;
A = eye(n,n);  % creates an identity matrix size n x n.
a = 0.1;
for i = 1:n-1
    A(i,i+1) = a;
    A(i+1,i) = a;
end
```

__Challenge 2__: Calculate and plot the determinant of tridiagonal matrices from sizes 2 to 50 with $\alpha = 0.1$

This is very similar to above, but we need another for loop so that $n$ iterates from 2 to 50.

```octave
j = 2:50;
detv = [];
for n = j
    A = eye(n,n);
    a = 0.1;
    for i = 1:n-1
        A(i,i+1) = a;
        A(i+1,i) = a;
    end
    detv = [detv det(A)];
end
plot(j,detv)
```

__Challenge 3__: Now plot on the same graph the determinants for size 2 to 50 with $\alpha = 0.01,0.02\cdots 0.1$

Again, very similar to above, but we need a for loop that will iterate $a$

```octave

j = 2:50;
alpha = 0.01:0.01:0.1;
for a = alpha
    detv = [];
    for n = j
        A = eye(n,n);
        for i = 1:n-1
            A(i,i+1) = a;
            A(i+1,i) = a;
        end
        detv = [detv det(A)];
    end
    plot(j,detv,'color',)
    hold on
end
```
